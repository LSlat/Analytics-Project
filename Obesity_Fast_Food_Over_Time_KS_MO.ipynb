```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Import Obesity data
health_data = pd.read_csv('cleaned_data/obese_data_converted_09_14.csv')
health_data.head()
```

## Affect Fast Food has on Obesity Over Time

```python
# Create new column to show percent change between years 2009 and 2014
health_data['Percent Change 09-14'] = health_data['PCT_OBESE_ADULTS14']-health_data['PCT_OBESE_ADULTS09']
health_data.head()
```

```python
# Import restaurant data for KS and MO
restaurant_data = pd.read_csv('cleaned_data/cleaned_restaurant_data_KS_MO.csv')
restaurant_data.head()
```

```python
# Define x axis values and y axis values 
x_axis = health_data['Percent Change 09-14'].tolist()
y_axis = restaurant_data['% Cng Fast Food Restautant Count 09_14'].tolist()
```

```python
# Create scatter plot
plt.scatter(x_axis, y_axis, edgecolor='black', color='blue', linewidth=0.4, s=20)

# Format look of scatter plot (labels, title, gridlines, etc.)
plt.title("# Fast Food Restaurants vs. Obesity Rate\n" + "% Change 2009-2014")
plt.xlabel('% Change in # of Fast Food Restaurants')
plt.ylabel('% Change Obesity Rate')
plt.grid()

# Export Figure
plt.savefig('saved figs/% Cng FF_vs_Obesity_KS_MO.png', dpi=160)
```
