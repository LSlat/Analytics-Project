```python
# Import Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Import population data for KS And MO
pop_data = pd.read_csv("cleaned_data/cleaned_population_data_KS_MO.csv")
pop_data.head()
```

```python
# Import restaurant data
restaurant_data = pd.read_csv("cleaned_data/cleaned_restaurant_data_KS_MO.csv")
restaurant_data.head()
```

```python
# Define x axis values and y axis values 
x_axis = pop_data['Population Estimate, 2014'].tolist()
y_axis = restaurant_data['Fast Food Restaurant Count 2014'].tolist()
```

```python
# Create scatter plot
plt.scatter(x_axis, y_axis, edgecolor='black', color='blue', linewidth=0.4, s=20)

# Format look of scatter plot (labels, title, gridlines, etc.)
plt.title("KS & MO 2014\n" + "Population vs # Fast Food Restaurants")
plt.xlabel('Population')
plt.ylabel('# Fast Food Restaurants')
plt.grid()

# Export Figure
plt.savefig('saved figs/pop_vs_#FF_KS_MO.png', dpi=160)
```
