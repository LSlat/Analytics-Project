```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Import Obesity data
health_data = pd.read_csv('cleaned_data/obese_data_converted_09_14.csv')
health_data.head()
```

```python
# Import Restaurant Data
restaurant_data= pd.read_csv('cleaned_data/cleaned_restaurant_data_KS_MO.csv')
restaurant_data.head()
```

## Number Fast Food Restaurants vs. Obesity Rate

```python
# Define x axis values and y axis values 
y_axis = health_data['PCT_OBESE_ADULTS14'].tolist()
x_axis = restaurant_data['Fast Food Restaurant Count 2014'].tolist()
```

```python
# Create scatter plot
plt.scatter(x_axis, y_axis, edgecolor='black', color='blue', linewidth=0.4, s=20)

# Format look of scatter plot (labels, title, gridlines, etc.)
plt.title("KS & MO 2014\n" + "# Fast Food Restaurants vs. Obesity Rate")
plt.ylabel('Obesity Rate (%)')
plt.xlabel('# Fast Food Restaurants')
plt.grid()

# Export Figure
plt.savefig('saved figs/#FF_vs_Obesity_KS_MO.png', dpi=160)
```

```python
# Define x axis values and y axis values 
y_axis = health_data['PCT_OBESE_ADULTS14'].tolist()
x_axis = restaurant_data['Fast Food Restaurant Count 2014 (per thou)'].tolist()
```

```python
# Create scatter plot
plt.scatter(x_axis, y_axis, edgecolor='black', color='blue', linewidth=0.4, s=20)

# Format look of scatter plot (labels, title, gridlines, etc.)
plt.title("KS & MO 2014\n" + "# Fast Food Restaurants (per thou) vs. Obesity Rate")
plt.ylabel('Obesity Rate (%)')
plt.xlabel('# Fast Food Restaurants (per thou.)')
plt.grid()

# Export Figure
plt.savefig('saved figs/#FF_per_thousand_vs_Obesity_KS_MO.png', dpi=160)
```
