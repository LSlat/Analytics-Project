```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Import Obesity data
health_data = pd.read_csv('cleaned_data/obese_data_converted_09_14.csv')
health_data.head()
```

## Affect Number of Grocery Stores has on Obesity Over Time

```python
# Create new column to show percent change between years 2009 and 2014
health_data['Percent Change 09-14'] = health_data['PCT_OBESE_ADULTS14']-health_data['PCT_OBESE_ADULTS09']
health_data.head()
```

```python
# import store data for KS and MO
store_data = pd.read_csv('cleaned_data/cleaned_store_data_KS_MO.csv')
store_data.head()
```

```python
# Define x axis values and y axis values 
x_axis = health_data['Percent Change 09-14'].tolist()
y_axis = store_data['% Chg Grocery Store Count 09_14'].tolist()
```

```python
# Create scatter plot
plt.scatter(x_axis, y_axis, edgecolor='black', color='blue', linewidth=0.4, s=20)

# Format look of scatter plot (labels, title, gridlines, etc.)
plt.title("# Grocery Stores vs. Obesity Rate\n" + "% Change 2009-2014")
plt.xlabel('% Change in # of Grocery Stores')
plt.ylabel('% Change Obesity Rate')
plt.grid()

# Export Figure
plt.savefig('saved figs/% Cng Groc Store_vs_Obesity Rate_KS_MO.png', dpi=160)
```
